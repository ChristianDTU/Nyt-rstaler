```python
import numpy as np
```

Queen (0) accuracy:

```python
p_hat = 0.7175
n = 1643 # number of sentences in the test set

# Beregner 95% konfidensinterval
std_hat = np.sqrt((p_hat*(1-p_hat)/n))
# fra standardnormalfordeling
z_kvartil = 1.96

CI = p_hat + np.array([-1,1]) * z_kvartil * std_hat
print(f"The 95%-confidence interval for the Queen's accuracy score is: {CI}")
```
